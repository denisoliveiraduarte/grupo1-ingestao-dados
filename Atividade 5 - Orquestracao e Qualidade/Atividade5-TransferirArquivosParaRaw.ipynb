```python

import sys
from awsglue.utils import getResolvedOptions
from pyspark.context import SparkContext
from awsglue.context import GlueContext
from awsglue.job import Job
  
sc = SparkContext.getOrCreate()
glueContext = GlueContext(sc)
spark = glueContext.spark_session
job = Job(glueContext)
import boto3
bucketNameSource = "atividade5-source"
bucketNameRaw = "atividade5-raw"
s3 = boto3.resource('s3')
bucketSource = s3.Bucket(bucketNameSource)
pastaBancos = "bancos"
pastaEmpregados = "empregados"
pastaReclamacoes = "reclamacoes"

for obj in bucketSource.objects.filter(Prefix=pastaBancos):
    source_filename = (obj.key).split('/')[-1]
    copy_source = {
        'Bucket': bucketNameSource,
        'Key': obj.key
    }
    target_filename = "{}/{}".format(pastaBancos, source_filename)
    if source_filename != '':
        s3.meta.client.copy(copy_source, bucketNameRaw, target_filename)
        
for obj in bucketSource.objects.filter(Prefix=pastaEmpregados):
    source_filename = (obj.key).split('/')[-1]
    copy_source = {
        'Bucket': bucketNameSource,
        'Key': obj.key
    }
    target_filename = "{}/{}".format(pastaEmpregados, source_filename)
    if source_filename != '':
        s3.meta.client.copy(copy_source, bucketNameRaw, target_filename)
        
for obj in bucketSource.objects.filter(Prefix=pastaReclamacoes):
    source_filename = (obj.key).split('/')[-1]
    copy_source = {
        'Bucket': bucketNameSource,
        'Key': obj.key
    }
    target_filename = "{}/{}".format(pastaReclamacoes, source_filename)
    if source_filename != '':
        s3.meta.client.copy(copy_source, bucketNameRaw, target_filename)
    
job.commit()
```
